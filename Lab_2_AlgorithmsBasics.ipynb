```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
print(func2(3, 4))
# it is an iterative function which return the power of the number
import time
start=time.time()
func2(2, 4)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10

   return result
```

```python
# Sample function call: func3(1234)
print(func3(1234))
# this function is used to return the sum of digit of the given number
import time
start=time.time()
func3(1234)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
# the function use to compare two list and return the number of time duplicate element occurs in the list
import time
start=time.time()
func4([1, 2, 3, 4], [2, 4, 6, 8])
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
print(func5(16))
# this function returns the squareroot of the perfect square number otherwise it returns none.
import time
start=time.time()
func5(16)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
print(func12([1, 2, 3, 4, 5], 2))
# this function used to perform circular rotation of the element of list by k times as passed in the function
import time
start=time.time()
func12([1, 2, 3, 4, 5], 2)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in arr:
    sum=sum+i

  return sum
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even=0
  odd=0
  for i in arr:
    if i%2==0:
      even=even+1
    else:
      odd=odd+1
  print(f"No of odd={odd},No of even={even}")
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,target):
  for i in arr:
    if i==target:
      return True

  return False
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
array = [1, 3, 5, 7, 9]
if linear_Search(array,3):
  print("Element found")
else:
  print("Element not found")

import time
start=time.time()
linear_Search(array,3)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def binary_search(arr, target):
    low = 0
    high=len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9,15,16,14,45]
if binary_search(arr,2)!=-1:
  print("Element found")
else:
  print("Element not found")

import time
start=time.time()
binary_search(arr,2)
end=time.time()
elapsedtime=(end-start)*1000
print(elapsedtime)
```
